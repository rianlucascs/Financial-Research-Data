# Cálculo do Drawdown Atual - Grupo de Ativos

Calcula o drawdown atual de cada ativo e armazena em um DataFrame.

```python
import yfinance as yf
import pandas as pd
from datetime import datetime
```

```python
tickers = [
    "ABEV3.SA", "ALOS3.SA", "AXIA6.SA", "B3SA3.SA", "BBDC3.SA", "BBDC4.SA",
    "BPAC11.SA", "BRAP4.SA", "BRAV3.SA", "CPFE3.SA", "CPLE3.SA", "CURY3.SA",
    "EGIE3.SA", "EMBJ3.SA", "ENEV3.SA", "ENGI11.SA", "EQTL3.SA", "FLRY3.SA",
    "GGBR4.SA", "GOAU4.SA", "IGTI11.SA", "IRBR3.SA", "ISAE4.SA", "ITSA4.SA",
    "ITUB4.SA", "KLBN11.SA", "MOTV3.SA", "MRVE3.SA", "MULT3.SA", "PRIO3.SA",
    "PSSA3.SA", "RADL3.SA", "RENT3.SA", "SANB11.SA", "SBSP3.SA", "SLCE3.SA",
    "SUZB3.SA", "TAEE11.SA", "TIMS3.SA", "UGPA3.SA", "USIM5.SA", "VALE3.SA",
    "VBBR3.SA", "VIVT3.SA", "WEGE3.SA",

    '^BVSP'
]
```

```python
# Baixar dados históricos
print("Baixando dados...")
data = yf.download(tickers, period="5y", progress=False)['Close']
print(f"✓ {data.shape[0]} dias, {data.shape[1]} ativos")
```

```python
# Calcular o drawdown atual de cada ativo
drawdown_atual = []

for ticker in data.columns:
    prices = data[ticker].dropna()
    
    # Máximo histórico
    max_price = prices.max()
    
    # Preço atual (último)
    current_price = prices.iloc[-1]
    
    # Drawdown atual = (preço_atual - máximo) / máximo * 100
    dd = ((current_price - max_price) / max_price) * 100
    
    drawdown_atual.append(dd)

# Criar DataFrame com os tickers e o drawdown atual
df = pd.DataFrame({
    'Ticker': data.columns,
    'Drawdown_Atual_%': drawdown_atual
})

# Ordenar do pior para o melhor
df = df.sort_values('Drawdown_Atual_%').reset_index(drop=True)

df
```
