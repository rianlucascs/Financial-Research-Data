```python
import yfinance as yf
import pandas as pd
import plotly.express as px

# =====================
# BAIXAR DADOS
# =====================

# Ibovespa
ibov = yf.download("^BVSP", start="2010-01-01", progress=False)["Close"]

# Dólar comercial (USD/BRL)
usd_brl = yf.download("USDBRL=X", start="2010-01-01", progress=False)["Close"]

# =====================
# UNIR EM UM DATAFRAME
# =====================

df = pd.concat([ibov, usd_brl], axis=1)
df.columns = ["Ibov", "USD_BRL"]

# remover NaNs
df = df.dropna()

# =====================
# CALCULAR Ibovespa EM DÓLARES
# =====================

df["Ibov_USD"] = df["Ibov"] / df["USD_BRL"]

# =====================
# NORMALIZAR BASE 100
# =====================

df_norm = df / df.iloc[0] * 100

# =====================
# PLOT
# =====================

fig = px.line(
    df_norm.reset_index(),
    x="Date",
    y=["Ibov", "Ibov_USD", "USD_BRL"],
    template="plotly_white",
    height=650,
    title="Ibovespa vs Ibovespa em USD vs Dólar (Base 100)"
)

fig.update_layout(
    title=dict(x=0.5),
    xaxis_title="Data",
    yaxis_title="Base 100",
    legend_title=None
)

fig.update_xaxes(showgrid=True)
fig.update_yaxes(showgrid=True)

fig.show()
```
