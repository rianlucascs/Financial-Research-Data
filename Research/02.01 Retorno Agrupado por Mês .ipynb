```python
import warnings
warnings.filterwarnings('ignore')
from yfinance import download
import plotly.express as px
```

```python
param = {
    'ticker': 'klbn4.sa',
    'interval': '1d',
    'period': 'max',
}

df = download(param['ticker'], period=param['period'], interval=param['interval'], progress=False, auto_adjust=False).droplevel(1, axis=1)

df['ret'] = df['Close'].pct_change()
df['mes'] = df.index.month
df['ano'] = df.index.year

tabela = (
    df.groupby(['ano', 'mes'])[['ret']]
    .sum()
    .reset_index()
    .pivot(index='ano', columns='mes', values='ret')
)
```

```python
tabela
```

```python
fig = px.imshow(
    tabela,
    text_auto='.5f',            # anotações com 5 casas
    color_continuous_scale='RdYlGn',
    color_continuous_midpoint=0,  # centro em zero
    aspect='auto'
)

fig.update_layout(
    title=dict(
        text=f'Retorno Ano e Mês - {param["ticker"].upper()}',
        x=0.5,
        xanchor='center',
        font=dict(size=18)
    ),
    coloraxis_colorbar=dict(
        title='Retorno Médio',
        len=0.8
    ),
    xaxis=dict(
        title='',
        tickangle=0
    ),
    yaxis=dict(
        title=''
    ),
    height=500,
    template='plotly_white',
)

# Rodapé
fig.add_annotation(
    text=f"<i>Intervalo do retorno: {param['interval']} | "
         f"Período: {param['period']} | Fonte: Yahoo Finance</i>",
    xref='paper',
    yref='paper',
    x=0.5,
    y=-0.12,
    showarrow=False,
    font=dict(
        size=11,
        color='gray'
    ),
    xanchor='center'
)

fig.show()
```
