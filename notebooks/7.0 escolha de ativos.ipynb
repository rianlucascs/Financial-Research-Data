```python
from yfinance import download
from pandas import DataFrame, concat
import warnings
warnings.filterwarnings("ignore")
```

```python
tickers = ['ABEV3.SA','AURE3.SA','B3SA3.SA','BBDC3.SA','BBDC4.SA','CPFE3.SA','EGIE3.SA','EMBJ3.SA',
           'ENGI11.SA','EQTL3.SA','FLRY3.SA','HYPE3.SA','IGTI11.SA','ISAE4.SA','ITSA4.SA','ITUB4.SA',
           'KLBN11.SA','MOTV3.SA','MULT3.SA','PSSA3.SA','RADL3.SA','SANB11.SA','SLCE3.SA','SUZB3.SA',
           'TAEE11.SA','TIMS3.SA','TOTS3.SA','VBBR3.SA','VIVT3.SA','WEGE3.SA']

df_data = download(tickers, period="10y", interval="1d", auto_adjust=False)
```

```python
faltantes = df_data['Adj Close'].isnull().sum()
faltantes = faltantes[faltantes > 0]
faltantes
```

```python
list(faltantes.index)
```

```python
# retorno mensal por ativo

df_ret = DataFrame()

exceto = ['AURE3.SA', 'EMBJ3.SA', 'IGTI11.SA', 'MOTV3.SA', 'VBBR3.SA']

for ticker in tickers:
    if not ticker in exceto:
        ret_mensal = df_data.xs(ticker, level='Ticker', axis=1)

        ret_mensal = ret_mensal.resample('M').agg(
            open_mes=('Open', 'first'),
            close_mes=('Close', 'last')
        )

        ret_mensal[ticker] = (
            ret_mensal['close_mes'] - ret_mensal['open_mes']
        )
        df_ret = concat([df_ret, ret_mensal[ticker]], axis=1)
```
