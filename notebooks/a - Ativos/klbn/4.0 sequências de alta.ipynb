```python
import warnings
warnings.filterwarnings("ignore")
from yfinance import download
```

```python
ticker = 'klbn4.sa'
```

```python
df = download(ticker, period="max", interval="1d", auto_adjust=False).droplevel(1, axis=1)
```

```python
df
```

```python
import pandas as pd

# Garantir datetime
df.index = pd.to_datetime(df.index)

# Primeiro dia do mês
primeiro_dia_mes = df.index.to_period('M').to_timestamp()

# Semana do mês
df['semana_mes'] = (
    ((df.index.day - 1 + primeiro_dia_mes.weekday) // 7) + 1
)
```

```python
import numpy as np
import pandas as pd

df.index = pd.to_datetime(df.index)

primeiro_dia_mes = df.index.to_period('M').to_timestamp()
weekday_inicio = primeiro_dia_mes.weekday

df['semana_mes'] = (
    ((df.index.day - 1 + weekday_inicio) // 7) + 1
)

# Limitar no máximo a 5 semanas
df['semana_mes'] = np.where(df['semana_mes'] > 5, 5, df['semana_mes'])
```

```python
df['semana_mes'].value_counts()
```

```python
df['ano_mes'] = df.index.to_period('M')
df['ano'] = df.index.year
df['mes'] = df.index.month
```

```python
df.index = pd.to_datetime(df.index)

mapa_dias = {
    0: 'Segunda-feira',
    1: 'Terça-feira',
    2: 'Quarta-feira',
    3: 'Quinta-feira',
    4: 'Sexta-feira',
    5: 'Sábado',
    6: 'Domingo'
}

df['dia_semana'] = df.index.dayofweek.map(mapa_dias)
```

```python
df.tail(50)
```

```python
df.loc[df['semana_mes'] == 5]
```

```python
df['ret'] = df['Close'].pct_change(1)
df.tail(3)
```

```python
df['ret_acum_semana'] = (
    df
    .sort_index()
    .groupby(['ano_mes', 'semana_mes'])['ret']
    .cumsum()
)
```

```python
df.tail(20)
```

```python
ret_ultimo = (
    df
    .groupby(['ano', 'mes', 'semana_mes'])['ret_acum_semana']
    .last()
    .reset_index()
)


```

```python
ret_ultimo
```

```python
for i in range(1, 6):
    perc_pos = len(ret_ultimo.loc[(ret_ultimo['semana_mes'] == i) & (ret_ultimo['ret_acum_semana'] > 0)]) / len(ret_ultimo.loc[ret_ultimo['semana_mes'] == i])
    print(f'{i} {perc_pos:.2%}')
```

```python
for i in range(1, 6):      
    perc_neg = len(ret_ultimo.loc[(ret_ultimo['semana_mes'] == i) & (ret_ultimo['ret_acum_semana'] < 0)]) / len(ret_ultimo.loc[ret_ultimo['semana_mes'] == i])
    print(f'{i} {perc_neg:.2%}')
```

```python
import plotly.express as px

heat = ret_ultimo.pivot_table(
    values='ret_acum_semana',
    index='ano',
    columns='semana_mes',
    aggfunc='mean'
)

fig = px.imshow(
    heat,
    aspect='auto',
    color_continuous_scale='RdBu',
    title='Heatmap – Retorno médio por semana do mês'
)

fig.update_layout(
    width=1000,
    height=700
)

fig.show()

```

```python
fig = px.box(
    ret_ultimo,
    x='semana_mes',
    y='ret_acum_semana',
    title='Distribuição do retorno acumulado por semana do mês'
)

fig.update_layout(
    width=1000,
    height=600
)

fig.show()
```
