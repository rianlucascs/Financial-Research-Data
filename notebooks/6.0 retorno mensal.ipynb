```python
from yfinance import download
df_data = download("vale3.sa", period="10y", interval="1d", auto_adjust=False).droplevel(1, axis=1)
```

```python
df_data
```

```python
ret_mensal_1 = df_data.resample('M').agg(
    open_mes=('Open', 'first'),
    close_mes=('Close', 'last')
)

ret_mensal_1['retorno_mensal'] = (
    ret_mensal_1['close_mes'] - ret_mensal_1['open_mes']
)
```

```python
ret_mensal_1.tail(30)
```

```python
ret_mensal_2 = df_data['Close'].pct_change(1).resample('M').sum()
```
